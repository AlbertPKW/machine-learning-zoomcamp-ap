
# Resampling to Boost Recall & F1 (while keeping Precision reasonable)

This notebook compares multiple strategies for imbalanced classification:

1. **Baseline (no resampling)**  
2. **SMOTE** (oversampling minority classes)  
3. **SMOTE + Tomek** (oversample + clean borderline/noisy samples)  
4. **Class Weights / Sample Weights** (penalize mistakes on minority classes)  
5. **(Binary only)** Threshold tuning to maximize F1 without tanking precision

> It is designed to handle **binary and multiclass** targets.  
> You only need to point `CSV_PATH` and `TARGET_COL` in the next cell.


```python

# === EDIT THESE ===
CSV_PATH = "train_LZdllcl.csv"  # set your data path (CSV)
TARGET_COL = 'is_promoted'                          # if None, will try to infer (last column)
ID_COLS = []                               # optional id-like columns to drop
RANDOM_STATE = 42

# Optional: downsample columns for speed when experimenting
# KEEP_COLUMNS = None        # e.g., ['f1','f2','f3', 'target']
KEEP_COLUMNS = None
```

```python

import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

# plotting
import matplotlib.pyplot as plt

# modeling
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import (
    classification_report, confusion_matrix, ConfusionMatrixDisplay,
    precision_recall_fscore_support, accuracy_score, f1_score, precision_score, recall_score,
    average_precision_score, roc_auc_score, roc_curve, precision_recall_curve
)
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.utils.class_weight import compute_class_weight

# models: prefer XGBoost if available, fallback to RandomForest
try:
    from xgboost import XGBClassifier
    HAVE_XGB = True
except Exception as e:
    from sklearn.ensemble import RandomForestClassifier
    HAVE_XGB = False

# imbalanced-learn (resampling)
HAVE_IMBLEARN = True
try:
    from imblearn.over_sampling import SMOTE
    from imblearn.combine import SMOTETomek
    from imblearn.pipeline import Pipeline as ImbPipeline
except Exception as e:
    HAVE_IMBLEARN = False
    print("imblearn not available. SMOTE/SMOTETomek steps will be skipped.",
          "\nInstall: pip install imbalanced-learn")
```

```python

# Load CSV
df = pd.read_csv(CSV_PATH)

# Optionally keep subset of columns
if KEEP_COLUMNS is not None:
    df = df[KEEP_COLUMNS]

# Infer target if needed
if TARGET_COL is None:
    TARGET_COL = df.columns[-1]

# Drop ID columns if present
for c in ID_COLS:
    if c in df.columns:
        df = df.drop(columns=[c])

# Separate X / y
y = df[TARGET_COL]
X = df.drop(columns=[TARGET_COL])

# Identify types
cat_cols = X.select_dtypes(include=["object", "category"]).columns.tolist()
num_cols = X.select_dtypes(exclude=["object", "category"]).columns.tolist()

print("Shape:", X.shape, "Target:", TARGET_COL)
print("Numeric cols:", len(num_cols), "| Categorical cols:", len(cat_cols))
print("Target distribution:\n", y.value_counts(normalize=True).round(3))
is_binary = y.nunique() == 2
classes_sorted = sorted(y.unique())
```

```python

# Column transformer: numerical -> impute median; categorical -> impute missing + one-hot
numeric_tf = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median"))
])

categorical_tf = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_tf, num_cols),
        ("cat", categorical_tf, cat_cols)
    ]
)
```

```python

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_STATE, stratify=y
)

print("Train size:", X_train.shape, " Test size:", X_test.shape)
```

```python

def build_model():
    if HAVE_XGB:
        # sensible default XGB settings
        model = XGBClassifier(
            n_estimators=300,
            learning_rate=0.1,
            max_depth=6,
            subsample=0.8,
            colsample_bytree=0.8,
            tree_method='hist',
            random_state=RANDOM_STATE,
            eval_metric='logloss'
        )
    else:
        # fallback RF
        from sklearn.ensemble import RandomForestClassifier
        model = RandomForestClassifier(
            n_estimators=400,
            max_depth=None,
            min_samples_split=5,
            min_samples_leaf=2,
            class_weight=None,
            random_state=RANDOM_STATE,
            n_jobs=-1
        )
    return model
```

```python

def compute_metrics(y_true, y_pred, average='binary'):
    # for multiclass, use 'macro' averaging
    if average == 'binary':
        avg = 'binary'
    else:
        avg = 'macro'
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, average=avg, zero_division=0)
    rec = recall_score(y_true, y_pred, average=avg, zero_division=0)
    f1 = f1_score(y_true, y_pred, average=avg, zero_division=0)
    return acc, prec, rec, f1
```

```python

baseline_pipe = Pipeline(steps=[
    ("prep", preprocessor),
    ("clf", build_model())
])

baseline_pipe.fit(X_train, y_train)
y_pred_base = baseline_pipe.predict(X_test)

avg_kind = 'binary' if is_binary else 'macro'
base_metrics = compute_metrics(y_test, y_pred_base, average=avg_kind)

print("=== Baseline (no resampling) ===")
print(classification_report(y_test, y_pred_base, zero_division=0))
cm = confusion_matrix(y_test, y_pred_base, labels=classes_sorted)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=classes_sorted)
disp.plot()
plt.title("Confusion Matrix - Baseline")
plt.show()
```

```python

smote_metrics = None

if HAVE_IMBLEARN:
    # Sampling strategy: balance up to the majority, but you can try a softer target, e.g. 0.7
    smote = SMOTE(random_state=RANDOM_STATE, sampling_strategy='not majority' if not is_binary else 'auto')

    smote_pipe = ImbPipeline(steps=[
        ("prep", preprocessor),
        ("smote", smote),
        ("clf", build_model())
    ])
    smote_pipe.fit(X_train, y_train)
    y_pred_smote = smote_pipe.predict(X_test)
    smote_metrics = compute_metrics(y_test, y_pred_smote, average=avg_kind)

    print("=== SMOTE ===")
    print(classification_report(y_test, y_pred_smote, zero_division=0))
    cm = confusion_matrix(y_test, y_pred_smote, labels=classes_sorted)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=classes_sorted)
    disp.plot()
    plt.title("Confusion Matrix - SMOTE")
    plt.show()
else:
    print("SMOTE skipped (imblearn not installed).")
```

```python

smt_metrics = None

if HAVE_IMBLEARN:
    smt = SMOTETomek(random_state=RANDOM_STATE)
    smt_pipe = ImbPipeline(steps=[
        ("prep", preprocessor),
        ("smt", smt),
        ("clf", build_model())
    ])
    smt_pipe.fit(X_train, y_train)
    y_pred_smt = smt_pipe.predict(X_test)
    smt_metrics = compute_metrics(y_test, y_pred_smt, average=avg_kind)

    print("=== SMOTE + Tomek ===")
    print(classification_report(y_test, y_pred_smt, zero_division=0))
    cm = confusion_matrix(y_test, y_pred_smt, labels=classes_sorted)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=classes_sorted)
    disp.plot()
    plt.title("Confusion Matrix - SMOTE+Tomek")
    plt.show()
else:
    print("SMOTE+Tomek skipped (imblearn not installed).")
```

```python

# We'll compute per-class weights and pass as sample_weight into fit.
# Works with XGB (fit has sample_weight) and RF (fit has sample_weight).
weights_metrics = None

# compute balanced weights
classes = np.unique(y_train)
if len(classes) > 1:
    cw = compute_class_weight(class_weight='balanced', classes=classes, y=y_train)
    class_to_w = {c:w for c,w in zip(classes, cw)}
    sw = y_train.map(class_to_w)

    weights_pipe = Pipeline(steps=[
        ("prep", preprocessor),
        ("clf", build_model())
    ])
    weights_pipe.fit(X_train, y_train, clf__sample_weight=sw.values)
    y_pred_w = weights_pipe.predict(X_test)
    weights_metrics = compute_metrics(y_test, y_pred_w, average=avg_kind)

    print("=== Class/Sample Weights ===")
    print(classification_report(y_test, y_pred_w, zero_division=0))
    cm = confusion_matrix(y_test, y_pred_w, labels=classes_sorted)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=classes_sorted)
    disp.plot()
    plt.title("Confusion Matrix - Class/Sample Weights")
    plt.show()
else:
    print("Class weights skipped: single-class target in training.")
```

```python

thresh_metrics = None
best_thresh = None

if is_binary:
    # We'll use the best performing of the above pipelines for probabilities, fallback to baseline
    candidate_pipes = []
    labels = []
    candidate_pipes.append(("Baseline", baseline_pipe))
    if HAVE_IMBLEARN:
        candidate_pipes.append(("SMOTE", smote_pipe))
        candidate_pipes.append(("SMOTE+Tomek", smt_pipe))
    candidate_pipes.append(("Weights", weights_pipe))

    # pick the one with highest F1 so far
    scores = []
    for name, pipe in candidate_pipes:
        y_pred = pipe.predict(X_test)
        acc, pr, rc, f1 = compute_metrics(y_test, y_pred, average='binary')
        scores.append((f1, name, pipe))
    scores.sort(reverse=True, key=lambda x: x[0])
    best_name, best_pipe = scores[0][1], scores[0][2]

    # probas for positive class
    y_proba = best_pipe.predict_proba(X_test)[:, 1]

    # sweep thresholds and pick best F1 with a constraint on precision (>= 0.65 by default; edit as needed)
    thresholds = np.linspace(0.1, 0.9, 41)
    best = (-1, 0.5, 0, 0)  # (F1, thresh, precision, recall)
    for t in thresholds:
        y_hat = (y_proba >= t).astype(int)
        p = precision_score(y_test, y_hat, zero_division=0)
        r = recall_score(y_test, y_hat, zero_division=0)
        f1 = f1_score(y_test, y_hat, zero_division=0)
        # enforce minimal precision to avoid drastic degradation
        if p >= 0.65 and f1 > best[0]:
            best = (f1, t, p, r)

    best_f1, best_thresh, best_p, best_r = best
    if best_f1 > 0:
        y_thresh = (y_proba >= best_thresh).astype(int)
        thresh_metrics = (accuracy_score(y_test, y_thresh),
                          best_p, best_r, best_f1)
        print(f"=== Threshold Tuning on best pipeline [{best_name}] ===")
        print(f"Best threshold = {best_thresh:.2f} | Precision={best_p:.3f} Recall={best_r:.3f} F1={best_f1:.3f}")

        # PR Curve
        prec, reca, th = precision_recall_curve(y_test, y_proba)
        plt.plot(reca, prec)
        plt.xlabel("Recall")
        plt.ylabel("Precision")
        plt.title("Precision-Recall Curve (Best Pipeline)")
        plt.show()
    else:
        print("No threshold improved F1 with required precision constraint.")
else:
    print("Threshold tuning skipped: target is multiclass.")
```

```python

rows = []
rows.append(("Baseline",) + base_metrics)

if smote_metrics is not None:
    rows.append(("SMOTE",) + smote_metrics)
if smt_metrics is not None:
    rows.append(("SMOTE+Tomek",) + smt_metrics)
if weights_metrics is not None:
    rows.append(("Weights",) + weights_metrics)
if thresh_metrics is not None:
    rows.append((f"BestThresh({best_thresh:.2f})",) + thresh_metrics)

res_df = pd.DataFrame(rows, columns=["Method", "Accuracy", "Precision", "Recall", "F1"])
display(res_df.sort_values(by="F1", ascending=False).reset_index(drop=True))

# Bar charts: one figure per metric (no custom colors)
for metric in ["Precision", "Recall", "F1"]:
    plt.figure()
    plt.bar(res_df["Method"], res_df[metric])
    plt.ylabel(metric)
    plt.title(f"{metric} by Method")
    plt.xticks(rotation=20)
    plt.tight_layout()
    plt.show()
```


## Notes & Tips
- If precision drops too much after SMOTE, try a **softer `sampling_strategy`** (e.g., `0.5`) or prefer **SMOTE+Tomek**.  
- For **XGBoost** (binary), instead of resampling you can also tune `scale_pos_weight = (#neg / #pos)` when not using resampling.  
- For **multiclass**, `SMOTE` supports strategy `'not majority'`; precision/recall are reported as **macro** averages.  
- Use **threshold tuning** to recover precision or recall: you can choose a minimum precision constraint in the code.  
- Consider **cross-validation** once the pipeline is stable to reduce variance in estimates.
