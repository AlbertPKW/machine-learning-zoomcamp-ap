### Train-Test Split

```python
dfy = df.is_promoted # model output
dfX = df.drop(['is_promoted', 'employee_id'], axis=1) # Model Inputs
dfX.head()
```

```python
# Split both Inputs (X) and Output (y) into training set (70%) and testing set (30%)
X_train, X_test, y_train, y_test = train_test_split(dfX, dfy, test_size=0.3, random_state=2)
```

### Data Imputation

```python
#let's create an arbitrary value imputer
imputer = ArbitraryNumberImputer(
    arbitrary_number=3, variables=['previous_year_rating'])

imputer.fit(X_train)
```

```python
# transform the data
X_train = imputer.transform(X_train)
X_test = imputer.transform(X_test)
```

```python
X_train.head()
```

```python
X_train.isnull().mean()
```

### Categorical Data Encoding

```python
text_columns = ['department', 'education', 'gender', 'recruitment_channel']
```

```python
# let's create the encoder
ordinal_enc = OrdinalEncoder(
    encoding_method='ordered',
    variables=text_columns
)
```

```python
ordinal_enc.fit(X_train, y_train)
```

```python
ordinal_enc.variables_
```

```python
ordinal_enc.encoder_dict_
```

```python
# let's transform train and test sets
X_train = ordinal_enc.transform(X_train)
X_test = ordinal_enc.transform(X_test)
```

### Build the XGBoost Models

#### Base Model

```python
# Build XGBoost Model and Evaluate accuracy
xgb = XGBClassifier(
    n_estimators=100,
    max_depth=5,
    random_state=2,
    n_jobs=-1
)

xgb.fit(X_train, y_train)
```

#### Fine-tuning by Grid Search

```python
# Determine hyperparameters to auto tune

param_grid = {
    'max_depth': [4, 5, 6],
    'min_child_weight': [1, 2, 5],
    'gamma': [0.1, 0.3, 1],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'learning_rate': [0.01, 0.05, 0.1],
    'n_estimators': [100, 200, 300]
}
```

```python
# Train a base model
xgb_base = XGBClassifier(random_state=2)

# Train a second XGBoost Model "xgb_gscv" with "best" hyperparameters
xgb_gscv = GridSearchCV(estimator=xgb_base, param_grid=param_grid, scoring='roc_auc', cv=5, n_jobs=-1)

# Fit the model to the training set
xgb_gscv = xgb_gscv.fit(X_train, y_train)
```

```python
print(xgb_gscv.best_score_)
print(xgb_gscv.best_params_)
```

```python
# Train a base XGBoost Model "xgb_gs"
xgb_gs = XGBClassifier(n_estimators=200,
                       min_child_weight=1,
                       learning_rate=0.1,
                       max_depth=4,
                       subsample=1.0,
                       colsample_bytree=0.6,
                       gamma=0.1,
                       random_state=2,
                       n_jobs=-1)

# Fit the model to the training set
xgb_gs.fit(X_train,y_train)
```

```python
# Train a RandomForestClassifier "xgb_fi" 
xgb_fi = XGBClassifier(n_estimators=200,
                       min_child_weight=1,
                       learning_rate=0.1,
                       max_depth=4,
                       subsample=1.0,
                       colsample_bytree=0.6,
                       gamma=0.1,
                       random_state=2,
                       n_jobs=-1)

# Fit the model to the training set
xgb_fi.fit(X_train_fi_xgb, y_train)
```

#### Adjusting Input data by Scaling

```python
# set up the scaler
scaler = MinMaxScaler()

# fit the scaler to the train set, it will learn the parameters
scaler.fit(X_train_fi_xgb)
```

```python
# transform train and test sets
X_train_scaled_xgb_fi = scaler.transform(X_train_fi_xgb)
X_test_scaled_xgb_fi = scaler.transform(X_test_fi_xgb)
```

```python
# Train a XGBoost Classifier Model "xgb_sc" 
xgb_sc = XGBClassifier(n_estimators=200,
                       min_child_weight=1,
                       learning_rate=0.1,
                       max_depth=4,
                       subsample=1.0,
                       colsample_bytree=0.6,
                       gamma=0.1,
                       random_state=2,
                       n_jobs=-1)

# Fit the model to the training set
xgb_sc.fit(X_train_scaled_xgb_fi, y_train)
```

### K-Fold Cross Validation

```python
dfX_bkup = dfX.copy()

imputer.fit(dfX_bkup)
dfX_bkup_imp = imputer.transform(dfX_bkup)

ordinal_enc.fit(dfX_bkup_imp, dfy)
dfX_bkup_imp = ordinal_enc.transform(dfX_bkup_imp)
```

```python
# Train a XGBoost Classifier Model "xgb_sc" 
xgb_kf = XGBClassifier(n_estimators=200,
                       min_child_weight=1,
                       learning_rate=0.1,
                       max_depth=4,
                       subsample=1.0,
                       colsample_bytree=0.6,
                       gamma=0.1,
                       random_state=2,
                       n_jobs=-1)
```

```python
# Evaluate the Model using K-folds Cross validation
xgb_kf_results  = cross_validate(xgb_kf, dfX_bkup_imp, dfy, scoring='roc_auc', cv=5, return_train_score = True, n_jobs=-1)
print('train_score: ', xgb_kf_results['train_score'])
print('test_score: ', xgb_kf_results['test_score'])
```

### Pipeline

```python
text_columns = ['department', 'education', 'gender', 'recruitment_channel']
```

```python
# Build Pipeline
xgb_pipeline = Pipeline(steps=[
    ("imputer", ArbitraryNumberImputer(
        arbitrary_number=3,
        variables=['previous_year_rating']
    )),
    
    ("ordinal_enc", OrdinalEncoder(
        encoding_method='ordered',
        variables=text_columns
    )),

    ("model", XGBClassifier(
        n_estimators=200,
        min_child_weight=1,
        learning_rate=0.1,
        max_depth=4,
        subsample=1.0,
        colsample_bytree=0.6,
        gamma=0.1,
        random_state=2,
        n_jobs=-1,
        tree_method='hist'
    ))
])
```

```python
# Fit the entire pipeline
xgb_pipeline.fit(X_train, y_train)
```

```python
y_pred_pl = xgb_pipeline.predict(X_test)
y_proba_pl = xgb_pipeline.predict_proba(X_test)[:,1]

# Various scores of XGBoost (Grid Search) 
print("Test Accuracy: %.3f" % metrics.accuracy_score(y_test, y_pred_pl))
print("Precision: %.3f" % metrics.precision_score(y_test, y_pred_pl))
print("Recall: %.3f" % metrics.recall_score(y_test, y_pred_pl))
print("F1 score: %.3f" % metrics.f1_score(y_test, y_pred_pl))
print("ROC-AUC: %.3f" % metrics.roc_auc_score(y_test, y_proba_pl))
```

```python
# Confusion Matrix
cnf_matrix = metrics.confusion_matrix(y_test, y_pred_pl)
cnf_matrix

ConfusionMatrixDisplay(confusion_matrix=cnf_matrix).plot()
```
